```python
import pyranges as pr
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(style='darkgrid')
```

```python
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = 15, 5
plt.rcParams['font.size'] = 12
```

```python
ENCFF136ECI_bed = 'H2AFZ_H1.ENCFF136ECI.hg19.bed'
ENCFF945OSD_bed = 'H2AFZ_H1.ENCFF945OSD.hg19.bed'
```

```python
data1 = pr.read_bed(ENCFF136ECI_bed, as_df=True)
data2 = pr.read_bed(ENCFF945OSD_bed, as_df=True)
```

```python
print('Количество пиков в {file}:'.format(file=ENCFF136ECI_bed))
print(data1.shape[0])
print('Количество пиков в {file}:'.format(file=ENCFF945OSD_bed))
print(data2.shape[0])
```

```python
data1
```

```python
data2
```

```python
data1['Length'] = data1['End'] - data1['Start']
data2['Length'] = data2['End'] - data2['Start']
```

```python
hist = sns.histplot(data1['Length'].values, bins=100)
hist.set_xlabel('Длина участка')
hist.set_title('Гистограмма длин участков эксперимента ENCFF136ECI')
plt.show()
```

```python
hist = sns.histplot(data2['Length'].values, bins=100)
hist.set_xlabel('Длина участка')
hist.set_title('Гистограмма длин участков эксперимента ENCFF945OSD')
plt.show()
```

```python
data1 = data1[data1['Length'] < 800]
data2 = data2[data2['Length'] < 400]
```

```python
hist = sns.histplot(data1['Length'].values, bins=100)
hist.set_xlabel('Длина участка')
hist.set_title('Гистограмма длин участков эксперимента ENCFF136ECI после фильтрации')
plt.show()
```

```python
hist = sns.histplot(data2['Length'].values, bins=20)
hist.set_xlabel('Длина участка')
hist.set_title('Гистограмма длин участков эксперимента ENCFF945OSD после фильтрации')
plt.show()
```

```python
print('Количество пиков из {file} после фильтрации:'.format(file=ENCFF136ECI_bed))
print(data1.shape[0])
print('Количество пиков из {file} после фильтрации:'.format(file=ENCFF945OSD_bed))
print(data2.shape[0])
```

```python
pr.from_dict(data1).to_bed('filter_' + ENCFF136ECI_bed)
```

```python
pr.from_dict(data2).to_bed('filter_' + ENCFF945OSD_bed)
```
